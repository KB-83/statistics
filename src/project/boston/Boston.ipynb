```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm

# 1

file_path = '../../../resources/project/boston.csv'

# Load
df = pd.read_csv(file_path)
df.drop_duplicates(inplace = True)


print(df);
print(df.describe())


# 2 : Relation between some variables and target
variables = ['RM', 'LSTAT', 'PTRATIO', 'CRIM']

# Set up the matplotlib figure
plt.figure(figsize=(16, 12))

for i, variable in enumerate(variables):
    plt.subplot(2, 2, i+1)
    sns.scatterplot(x=df[variable], y=df['MEDV'])
    plt.title(f'{variable} vs target')
    plt.xlabel(variable)
    plt.ylabel('target')

plt.tight_layout()
plt.show()


# 3:target variable is MDEV and other variables are prediction varialbes
# calculate the corralation map select predictor variables due to corr

target = 'MEDV'

corr_matrix = df.corr()

target_corr = corr_matrix[[target]].sort_values(by=target, ascending=False)

# Plot the heatmap for the correlation with the target variable
plt.figure(figsize=(8, 12))
sns.heatmap(target_corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation of Predictors with MEDV')
plt.show()


# 4:test and train

predictors = ['RM', 'LSTAT']
target = 'MEDV'

train_df, test_df = train_test_split(df, test_size=0.1, random_state=49)

X_train = train_df[predictors]
y_train = train_df[target]
X_test = test_df[predictors]
y_test = test_df[target]


model = LinearRegression()
model.fit(X_train, y_train)



# Part 2 : Simple Linear Regression

def fit_and_evaluate(predictor):
    X = df[[predictor]]
    y = df['MEDV']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=37)

    X_train_sm = sm.add_constant(X_train)
    model = sm.OLS(y_train, X_train_sm).fit()

    y_pred_train = model.predict(X_train_sm)
    y_pred_test = model.predict(sm.add_constant(X_test))

    RSS = ((y_train - y_pred_train) ** 2).sum()
    TSS = ((y_train - y_train.mean()) ** 2).sum()
    MSE_train = mean_squared_error(y_train, y_pred_train)
    R_squared = model.rsquared
    n = X_train_sm.shape[0]
    p = X_train_sm.shape[1] - 1
    Adjusted_R_squared = 1 - (1 - R_squared) * (n - 1) / (n - p - 1)
    p_value = model.pvalues[1]

    print(f"Model with {predictor}:")
    print(model.summary())
    print(f'RSS: {RSS:.2f}')
    print(f'TSS: {TSS}')
    print(f'MSE (Train): {MSE_train}')
    print(f'R Squared: {R_squared}')
    print(f'Adjusted R Squared: {Adjusted_R_squared}')
    print(f'p_value : {p_value}')
    print('\n')

    plt.figure(figsize=(10, 6))
    plt.scatter(y_test, y_pred_test, alpha=0.7, label='Actual')
    # plt.plot(X_test, y_pred_test, '--g', label='Regression Line')
    X_train_sorted = X_train.sort_values(by=predictor)
    y_pred_train_sorted = model.predict(sm.add_constant(X_train_sorted))
    # plt.plot(X_train_sorted, y_pred_train_sorted, '--r', label='Regression Line')
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()],'--g')
    plt.xlabel(predictor)
    plt.ylabel('MEDV')
    plt.title(f'Actual vs Predicted MEDV (Test Data) using {predictor}')
    plt.legend()
    plt.show()

    MSE_test = mean_squared_error(y_test, y_pred_test)
    R_squared_test = r2_score(y_test, y_pred_test)

    print(f'MSE (Test): {MSE_test:.2f}')
    print(f'R Squared (Test): {R_squared_test:.2f}')
    print('--------------------------------------')

def plot_regression_line(predictor):
    X = df[[predictor]]
    y = df['MEDV']

    X_sm = sm.add_constant(X)
    model = sm.OLS(y, X_sm).fit()

    plt.figure(figsize=(10, 6))
    plt.scatter(X, y, alpha=0.7, label='Data Points')
    plt.plot(X, model.predict(X_sm), '--r', label='Regression Line')
    plt.xlabel(predictor)
    plt.ylabel('MEDV')
    plt.title(f'Regression Line for {predictor} vs MEDV')
    plt.legend()
    plt.show()



predictors = ['CRIM','INDUS', 'RM','DIS','NOX','CHAS'] # TEST
for p in predictors:
  plot_regression_line(p)
fit_and_evaluate('RM')


# Part 3 : Complex Linear Regression

def fit_and_evaluate_multivariate(predictors):
    X = df[predictors]
    y = df['MEDV']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=37)

    X_train_sm = sm.add_constant(X_train)
    X_test_sm = sm.add_constant(X_test)

    model = sm.OLS(y_train, X_train_sm).fit()

    y_pred_train = model.predict(X_train_sm)
    y_pred_test = model.predict(X_test_sm)

    RSS = ((y_train - y_pred_train) ** 2).sum()
    TSS = ((y_train - y_train.mean()) ** 2).sum()
    MSE_train = mean_squared_error(y_train, y_pred_train)
    R_squared = model.rsquared
    n = X_train.shape[0]
    p = X_train.shape[1] - 1
    Adjusted_R_squared = 1 - (1 - R_squared) * (n - 1) / (n - p - 1)

    print(f"Model with predictors: {predictors}")
    print(model.summary())
    print(f'RSS: {RSS:.2f}')
    print(f'TSS: {TSS}')
    print(f'MSE (Train): {MSE_train}')
    print(f'R Squared: {R_squared}')
    print(f'Adjusted R Squared: {Adjusted_R_squared}')
    print('\n')

    plt.figure(figsize=(10, 6))
    plt.scatter(y_test, y_pred_test, alpha=0.7, label='Predicted vs Actual')
    plt.xlabel('Actual MEDV')
    plt.ylabel('Predicted MEDV')
    plt.title(f'Actual vs Predicted MEDV (Test Data) using {predictors}')
    plt.legend()
    plt.show()

    MSE_test = mean_squared_error(y_test, y_pred_test)
    R_squared_test = r2_score(y_test, y_pred_test)

    print(f'MSE (Test): {MSE_test:.2f}')
    print(f'R Squared (Test): {R_squared_test:.2f}')

    # Plot the distribution of residuals
    residuals = y_test - y_pred_test
    plt.figure(figsize=(10, 6))
    plt.hist(residuals, bins=30, alpha=0.7)
    plt.xlabel('Residuals')
    plt.ylabel('Frequency')
    plt.title('Distribution of Residuals')
    plt.show()

    print('--------------------------------------')


# Example: Using different sets of predictors
fit_and_evaluate_multivariate(['RM', 'LSTAT', 'NOX']) # IMPORTANT
fit_and_evaluate_multivariate(['CRIM', 'ZN', 'INDUS', 'CHAS'])
fit_and_evaluate_multivariate(['DIS', 'RAD', 'TAX', 'PTRATIO', 'B'])
fit_and_evaluate_multivariate(['INDUS','LSTAT','PTRATIO'])

```
